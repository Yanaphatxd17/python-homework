```python
class Student():
    def __init__(self, sid, cid, name, lname, age, gender):
        self.__sid = sid
        self.__name = name
        self.__lname = lname
        self.__age = age
        self.__gender = gender

        if len(cid) == 13:
            self.__cid = cid
        else:
            print("หมายเลขบัตรต้องมีความยาว 13 หลักเท่านั้น")
            self.__cid = ""

    def setAge(self, newAge):
        if self.__age < newAge:
            self.__age = newAge
        else:
            print("อายุต้องเพิ่มขึ้นมากกว่าเดิมเท่านั้น")

    def setCid(self, cid):
        if len(cid) == 13 and self.__cid == "":
            self.__cid = cid
        else:
            print("หมายเลขบัตรต้องมีความยาว 13 หลักเท่านั้น")
            self.__cid = ""

    def showInfo(self):
        print("""ชื่อ: {}
นามสกุล: {}
รหัสนักศึกษา: {}
นามสกุล: {}
รหัสนักศึกษา: {}
หมายเลขบัตรประชาชน: {}
อายุ: {}
        """.format(self.__name, self.__lname, self.__sid, self.__cid, self.__age))
```
