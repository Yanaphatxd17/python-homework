```python
class Person:
    def __init__(self, cid, name, lname):
        self.cid = cid
        self.name = name
        self.lname = lname
        
    def info(self):
        print("{}: {} {}".format(self.cid, self.name, self.lname))
```

```python
class Student(Person):
    def __init__(self, cid, name, lname):
        super().__init__(cid, name, lname)

    def helloTeacher(self):
        print("สวัสดีครับคุณครู")
```

```python
personl = Person("1234567891233", "Phisan", "Sookkhee")
personl.info()
```

```python
student = Student("1234567891233", "Student", "ComSci")
student.info()
student.helloTeacher()
```

```python
 class Student:
    def __init__ (self, name, student_id, gpa):
        self.name = name
        self.student_id = student_id
        self.gpa = gpa

    def show_info(self):
        print("{}: {} {}".format(self.name, self.student_id, self.gpa))

    def is_honor(self):
        return self.gpa >= 3.25
s1 = Student("สมชาย", "6501001", "3.50")
s1.show_info()
s2 = Student("สมหญิง", "6501002", "2.80")
s2.show_info()
```
