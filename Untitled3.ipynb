```python
from datetime import datetime

class DigitClock:
    def __init__(self):
        pass

    def showTime(self):
        now = datetime.now()
        print("เวลาปัจจุบัน : {}".format(now.strftime("%H:%M:%S")))
```

```python
from datetime import date
class Calendar:
    def __init__(self):
        pass

    def showDate(self):
        today = date.today()
        print("วันที่ปัจจุบัน : {}".format(today))
```

```python
class FutureClock(DigitClock, Calendar):
    def __init__(self):
        pass

    def show(self):
        print("----นี่คือนาฬิกาแหน่งอนาคต----")
        self.showDate()
        self.showTime()
        print("--------------------------")
```

```python
my_clock = FutureClock()
my_clock.show()
```
